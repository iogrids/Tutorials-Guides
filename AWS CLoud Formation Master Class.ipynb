```python
# Parameters common to cloudFormation Template

1. Tags                         - All the resources that you create using cloudFormation template will have the same tag. You can add upto 50 tags for each cloudFormation template
2. Permissions                  - You will have to assign an IAM role for this cloudFormation template
3. Notification options         - To send emails or SMS when certain things happen
4. Timeouts                     - To ensure that the cloudFormation fails after a specific amount of time.
5. Rollback on Failure          - Allows the cloudFormation to go back to its previous state if it fails
6. Stack Policy                 - To protect some resources in your cloudFormation template
```

```python
# CloudFormation building blocks. Any aws template will have the below mentioned components
# To easily create a cloudFormation template in sublime you can install a plugin called cform (https://github.com/beaknit/cform)

# Template components 

Resources: Resources are referenced as AWS::aws-product-name::data-type-name -> eg: AWS::Serverless::Function.
           There are 224 resources in AWS. 
           Documentation: https://docs.aws.amazon.com/AWSCloudFormation/latest/UserGuide/aws-template-resource-type-ref.html
           Ec2 instance documentation: https://docs.aws.amazon.com/AWSCloudFormation/latest/UserGuide/aws-properties-ec2-instance.html

Parameters: search for "aws parameters documentation" in Google for documentation
Mappings
Outputs
Conditions
Metadata

# Template helpers

References
Functions
```

```python
Parameters

search for "aws parameters documentation" in Google

# Documentation: https://docs.aws.amazon.com/AWSCloudFormation/latest/UserGuide/parameters-section-structure.html

# Parameters can have the following values

Type: 
    String
    Number 
    CommaDelimitedList
    List
    AWS Parameter - to catch invalid values by matching against exisiting values in your AWS account. eg I want my parameter to be a security group id. In this case only a security group within your aws account will be allowed for this cloudFormation
Description: You can give a descriotion to your parameters
Constraints: For example the parameter should be a number between 1 and 10
ConstraintDescription: If someone gets an error due to a constraint you may have to give why that error has occurred here
Min/MaxLength: For example you want to have a password and the minimum and maximum length can be mentioned here
Min/MaxValue: For example you can mention that your port number should be between 4000 to 40,000
Defaults: To provide defaults for users to use their cloudFormation templates
AllowedValues(array): allowed values for your paramaters
AllowedPattern(regexp): allowed pattern for your paramaters
NoEcho (Boolean): It wont be displayed to the end user
```

```python
# To create a simple ec2 instance using cloud formation
# Save the file as ec2.yaml
# The below code does not have any security groups or elastic ip

---
Resources:
  MyInstance:
    Type: AWS::EC2::Instance
    Properties:
      AvailabilityZone: us-east-1a
      ImageId: ami-a4c7edb2
      InstanceType: t2.micro
```

```python
# To create an ec2 instance with security group and elastic ip
# Security groups are for providing enable and disable port access
# Elastic ip is static ip which you are assigning to your aws instance

---
Resources:
  MyInstance:
    Type: AWS::EC2::Instance
    Properties:
      AvailabilityZone: us-east-1a
      ImageId: ami-a4c7edb2
      InstanceType: t2.micro
      SecurityGroups:
        - !Ref SSHSecurityGroup
        - !Ref ServerSecurityGroup

  # an elastic IP for our instance
  MyEIP:
    Type: AWS::EC2::EIP
    Properties:
      InstanceId: !Ref MyInstance

  # our EC2 security group
  SSHSecurityGroup:
    Type: AWS::EC2::SecurityGroup
    Properties:
      GroupDescription: Enable SSH access via port 22
      SecurityGroupIngress:
      - CidrIp: 0.0.0.0/0
        FromPort: 22
        IpProtocol: tcp
        ToPort: 22

  # our second EC2 security group
  ServerSecurityGroup:
    Type: AWS::EC2::SecurityGroup
    Properties:
      GroupDescription: allow connections from specified CIDR ranges
      SecurityGroupIngress:
      - IpProtocol: tcp
        FromPort: 80
        ToPort: 80
        CidrIp: 0.0.0.0/0
      - IpProtocol: tcp
        FromPort: 22
        ToPort: 22
        CidrIp: 192.168.1.1/32
```
